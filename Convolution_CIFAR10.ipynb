Class for callback and function for nn model

````
class callback(keras.callbacks.Callback):
  def on_epoch_end(self,epoch,logs={}):
    pass

def createModel():
  return keras.Sequential([
        keras.layers.Conv2D(32,(3,3),input_shape=(32,32,3),activation=tf.nn.relu),
        keras.layers.Conv2D(32,(3,3),activation=tf.nn.relu),
        keras.layers.MaxPooling2D(2,2),
        keras.layers.Dropout(0.25),
        keras.layers.Conv2D(64,(3,3),activation=tf.nn.relu),
        keras.layers.Conv2D(64,(3,3),activation=tf.nn.relu),
        keras.layers.MaxPooling2D(2,2),
        keras.layers.Dropout(0.25),
        keras.layers.Flatten(),
        keras.layers.Dense(512, activation=tf.nn.relu),
        keras.layers.Dense(10, activation=tf.nn.softmax)
    ])
def scale(images):
  return (images / 255.0)
````

Now we will load the data..

````
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
x_train = scale(x_train)
x_test = scale(x_test)
test_data= (x_test, y_test)
````

And Finally train the nn

````
model = createModel()
model.summary()
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",metrics=['accuracy'])
history = model.fit(x_train,y_train,epochs=25,validation_data=test_data,callbacks=[callback(),keras.callbacks.EarlyStopping()])
````
