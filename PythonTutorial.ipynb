```
class Quadrado():
  def __init__(self,lado):
    self.__lado = lado

  def get_area(self):
    return self.__lado ** 2
  
  def get_perimetro(self):
    return self.__lado * 4
  
  def __str__(self):
    return "Quadrado:[lado:{}]".format(self.__lado)

q1 = Quadrado(4)
print(q1.get_perimetro())
```

```
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```
X = np.array([np.random.randn(32,32,3) for x in range(0,5)]).reshape(X.shape[0],-1).T
Y = np.random.randint(low=0,high=2,size=(1,X.shape[1]))
M = X.shape[1]
```

```
W = np.zeros((X.shape[0],1))
B = 0

for x in range(0,1):
  #Forward propagate
  A = sigmoid(np.dot(W.T,X) + B)
  L = (1/M)*(np.sum(-1*(Y*np.log(A) + (1-Y)*(np.log(1-A)))))
  print(L)
  #Back propagate
  dz = A-Y
  dw = (1/M)*(np.dot(X,dz.T)) #dot devido a ser a soma remember my dude produto escalar
  db = (1/M)*np.sum(dz)

  learning_rate = 0.01
  W -= learning_rate*dw
  B -= learning_rate*db
```
