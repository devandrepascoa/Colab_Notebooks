```
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow import keras
from tensorflow.keras import models
```

```
(x_train,y_train),(x_test,y_test) = keras.datasets.cifar10.load_data()
img = x_test[0]

from scipy import misc
img = misc.ascent()
plt.imshow(img)
```

```
filter_conv = np.array([
          [-1,0,1],
          [-2,0,2],
          [-1,0,1]])
```

```

def convolution(image,filter_conv):
  img_w = image.shape[0]
  img_h = image.shape[1]
  result_conv = np.zeros((img_w-1,img_h-1))
  #devido filtro 3*3 -> imagem reduz 1 pixel a volta
  for x in range(1,img_w-1):
    for y in range(1,img_h-1):
      result = 0.0
      result += (filter_conv[0,0] * image[x-1,y-1])
      result += (filter_conv[1,0] * image[x,y-1])
      result += (filter_conv[2,0] * image[x+1,y-1])
      result += (filter_conv[0,1] * image[x-1,y])
      result += (filter_conv[1,1] * image[x,y])
      result += (filter_conv[2,1] * image[x+1,y])
      result += (filter_conv[0,2] * image[x-1,y+1])
      result += (filter_conv[1,2] * image[x,y+1])
      result += (filter_conv[2,2] * image[x+1,y+1])
      if result > 255.0: 
        result = 255.0
      if result < 0.0:
        result = 0.0
      result_conv[x,y] = result
  return result_conv
        
```

```
plt.gray()
plt.grid(False)
t_img = convolution(img,filter_conv)
plt.imshow(t_img)
```

```
def max_pool(image):
  img_w = image.shape[0]
  img_h = image.shape[1]
  result_conv = np.zeros((int(img_w/2),int(img_h/2)))
  #devido filtro 3*3 -> imagem reduz 1 pixel a volta
  for x in range(0,img_w-1,2):
    for y in range(0,img_h-1,2):
      result = max([image[x,y],image[x+1,y+1],image[x+1,y],image[x,y+1]])
      result_conv[(int(x/2)),int(y/2)] = result
  return result_conv
```

```
plt.gray()
plt.grid(False)
plt.imshow(max_pool(t_img))
plt.show()
plt.imshow(max_pool(max_pool(t_img)))
plt.show()
plt.imshow(max_pool(max_pool(max_pool(t_img))))
plt.show()
plt.imshow(max_pool(max_pool(max_pool(max_pool(t_img)))))
plt.show()
plt.imshow(max_pool(max_pool(max_pool(max_pool(max_pool(t_img))))))
```
